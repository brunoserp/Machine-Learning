A Dream Housing Finance Company é especializada em empréstimos imobiliários e busca automatizar o processo de elegibilidade para empréstimos com base nas informações fornecidas pelos clientes em formulários de solicitação online. A empresa deseja identificar diferentes segmentos de clientes elegíveis para empréstimos, para poder direcionar suas campanhas de forma mais eficaz.

Os dados têm informações sobre os clientes como gênero, estado civil, educação, número de dependentes, renda, valor do empréstimo e histórico de crédito, entre outros. O objetivo é usar árvore de decisão para classificar e avaliar a concessão do empréstimo com base nas características de cada tomador. 
    [Dados no Kaggle](https://www.kaggle.com/datasets/rishikeshkonapure/home-loan-approval?resource=download)

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for file in os.listdir():
    if file.endswith('csv'):
        if 'train' in file:
            df_train = pd.read_csv(file)
        else:
            df_test = pd.read_csv(file)
```

```python
# as colunas mais relevantes com NaN são LoanAmount e Credit_History, que são de difícil obtenção. Decidi apagar as linhas com NA:
df_train = df_train.dropna(how='any')
```

```python
# loan_status é o rótulo. Como ele se relaciona com as demais variáveis?
# primeiramente, é preciso substituir variáveis categóricas em dummies
cols_categoricas = ['Gender','Married','Education','Self_Employed','Property_Area','Loan_Status']
for col in cols_categoricas:
    colunas = df_train[col].drop_duplicates().tolist()
    if len(colunas) > 2:
        print(f'A coluna {col} tem mais de 2 valores: {colunas}')
    else:
        print(f'{col} tem os valores {colunas}')
```

```python
# não vou usar essas colunas pra facilitar a análise
df_train.drop(columns='Property_Area',inplace=True)
df_train.drop(columns='Dependents',inplace=True)
```

```python
# mapa de correlação
df_train['Gender'] = df_train['Gender'].map({'Male':0, 'Female':1})
df_train['Married'] = df_train['Married'].map({'Yes':1, 'No':0})
df_train['Education'] = df_train['Education'].map({'Graduate':1, 'Not Graduate':0})
df_train['Self_Employed'] = df_train['Self_Employed'].map({'Yes':1, 'No':0})
df_train['Loan_Status'] = df_train['Loan_Status'].map({'Y':1, 'N':0})
df_train_selecionados = df_train[['Gender', 'Married', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Loan_Status']]

correlation_matrix = df_train_selecionados.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True, fmt='.2f', 
            xticklabels=correlation_matrix.columns, yticklabels=correlation_matrix.columns, square=True)  

plt.tight_layout() 
plt.show()
```

```python
# a única variável com correlação relevante com loan_status é credit_history. Vou usá-la na árvore de decisão
```

```python
# transformando as colunas categóricas em dummies
df_test['Gender'] = df_test['Gender'].map({'Male':0, 'Female':1})
df_test['Married'] = df_test['Married'].map({'Yes':1, 'No':0})
df_test['Education'] = df_test['Education'].map({'Graduate':1, 'Not Graduate':0})
df_test['Self_Employed'] = df_test['Self_Employed'].map({'Yes':1, 'No':0})
df_test_selecionados = df_test[['Gender', 'Married', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History']]
```

```python
# árvore de decisão: estimar os valores limite para classificação de cada folha, gerando a estrutura para a árvore
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

x = df_train_selecionados.drop('Loan_Status', axis=1)  # Variáveis independentes
y = df_train_selecionados['Loan_Status']  # Variável dependente
# por mais que exista df_train e df_test, a df_test não tem rótulo. Por isso vou usar só a treino para treino e teste
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Avaliando o modelo
# acurácia: classificações certas/total
accuracy = accuracy_score(y_test, y_pred)
# matriz com VV, FV, FF, VF)
conf_matrix = confusion_matrix(y_test, y_pred)
# precisão: VV / VV + FV
# recall: VV / VV + FF
# f1: 2* (precisão*recall)/precisão+recall
class_report = classification_report(y_test, y_pred)

print(f'Acurácia: {accuracy}')
print(f'Matriz de Confusão:\n{conf_matrix}')
print(f'Relatório de Classificação:\n{class_report}')
```
